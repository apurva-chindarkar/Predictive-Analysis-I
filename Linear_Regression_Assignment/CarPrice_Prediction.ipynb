```python
# Drop rows with the Duplicate rows, if any.
car_price = car_price.drop_duplicates(subset = None, keep = "first")
car_price.head()
```

```python
# Creating a new column "Companyname" from "carname" by splitting the name and only retaining companyname of the car
car_price["Companyname"] = car_price["CarName"].str.split(' ').str[0]
car_price.head()
```

```python
# Dropping column "CarName" and "car_ID" since it is not required 
car_price.drop(['car_ID', 'CarName'], axis =1, inplace = True)
car_price.head()
```

```python
# Correcting spelling errors and getting all names in same format
car_price.Companyname = car_price["Companyname"].str.lower()
car_price["Companyname"] = car_price["Companyname"].str.replace('maxda','mazda')
car_price["Companyname"] = car_price["Companyname"].str.replace('porcshce','porsche')
car_price["Companyname"] = car_price["Companyname"].str.replace('toyouta','toyota')
car_price["Companyname"] = car_price["Companyname"].str.replace('vokswagen','volkswagen')
car_price["Companyname"] = car_price["Companyname"].str.replace('vw','volkswagen')
car_price.head()
```

After getting box-plots for all the categorical variables, we can observe some relationship between each of these variables and price. Since none of the categorical variables plotted above are ordinal, all the categorical variables with two levels need to be converted into "0" & "1" and variables with more than two states/levels will be processed by creating dummy variables. 

```python
# Binary mapping for categorical variables with two levels
car_price['fueltype'] = car_price['fueltype'].map({'gas': 1, 'diesel': 0})
car_price['aspiration'] = car_price['aspiration'].map({'std': 1, 'turbo': 0})
car_price['doornumber'] = car_price['doornumber'].map({'two': 1, 'four': 0})
car_price['enginelocation'] = car_price['enginelocation'].map({'front': 1, 'rear': 0})
car_price.head()
```

```python
#"cylindernumber" converted from strings to int by mapping words with numerical values
car_price['cylindernumber'] = car_price['cylindernumber'].map({"two": 2, "three": 3, "four": 4, "five": 5, "six": 6,"eight": 8, "twelve": 12})
car_price.head()
```

```python
# To confirm the datatype of categorical variables converted to int
car_price.info()
```

Dummy Variables will be created for catergorical variables with more than two levels. 

```python
# While creating dummy variables, first dummy variable and original varaible column will be dropped. 
car_df = pd.get_dummies((car_price), drop_first = True)
car_df.info()
```

Splitting the data into TRAINING AND TESTING sets...

```python
from sklearn.model_selection import train_test_split

# We specify this so that the train and test data set always have the same rows, respectively

df_train, df_test = train_test_split(car_df, train_size = 0.7, test_size = 0.3, random_state = 100)
```

Rescaling the features using MinMax scaling:

```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
```

```python
# Apply scaler() to all the columns of numeric variables
num_vars = ['symboling', 'wheelbase', 'carlength', 'carwidth', 'carheight', 
            'curbweight', 'cylindernumber', 'enginesize', 'boreratio', 'stroke',
            'compressionratio', 'horsepower', 'peakrpm', 'citympg', 'highwaympg', 'price']

df_train[num_vars] = scaler.fit_transform(df_train[num_vars])

df_train.head()
```

Dividing data into X and Y sets for building the model and using the **LinearRegression function from SciKit Learn** for model building 

```python
y_train = df_train.pop('price')
X_train = df_train
```

```python
# Importing RFE and LinearRegression
from sklearn.feature_selection import RFE
from sklearn.linear_model import LinearRegression
```

```python
# Running RFE with the output number of the variable equal to 20
lm = LinearRegression()
lm.fit(X_train, y_train)

rfe = RFE(lm, 20)             # running RFE
rfe = rfe.fit(X_train, y_train)
```
